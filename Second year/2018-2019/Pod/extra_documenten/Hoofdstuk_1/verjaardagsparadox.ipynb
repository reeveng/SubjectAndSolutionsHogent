# De verjaardagsparadox

```python
import numpy as np
import matplotlib.pyplot as plt
```

### De formule dat k personen allemaal een verschillende verjaardag hebben

```python
def allemaal_verschillend(k):
    voorwaardelijke_kansen = [(366-j)/365  for j in range(1, k+1)]  
    return np.prod(np.array(voorwaardelijke_kansen))
```

Verzamel alle waarden in tabel. Voeg een 'dummy' waarde toe vooraan 
zodat tabel[k] precies overeenkomt met de kans dat k personen allemaal
een verschillende verjaardag hebben.

```python
tabel = [-1] + [allemaal_verschillend(k) for k in range(1, 366)]
```

De waarden uit de tabel in de oplossingencursus

```python
tabel[5], tabel[10], tabel[15], tabel[20], tabel[21], tabel[22], \
tabel[23], tabel[24], tabel[30], tabel[40]
```

Een plot voor alle waarden van k

```python
plt.plot( range(1,366), tabel[1:366])
plt.hlines(y = 0.5, xmin = 1, xmax = 366)
plt.xlabel("Aantal personen")
plt.ylabel("Kans op allemaal verschillende verjaardag")
```

Een plot voor een kleinere range van k

```python
k_min = 20
k_max = 30
plt.plot( range(k_min,k_max+1), tabel[k_min:k_max + 1])
plt.hlines(y = 0.5, xmin = k_min, xmax = k_max + 1)
plt.xlabel("Aantal personen")
plt.ylabel("Kans op allemaal verschillende verjaardag")
```

## Simulatie
In plaats van de kans theoretisch te berekenen trachten we deze te benaderen aan de hand van een simulatie.

```python
def simulatie_verjaardagsparadox(aantal_personen, aantal_pogingen):
    allemaal_verschillend = []
    for _ in range(aantal_pogingen):
        verjaardagen = np.random.randint(0, 366, size = aantal_personen)
        unieke_verjaardagen = np.unique(verjaardagen)
        allemaal_verschillend.append(unieke_verjaardagen.shape[0] == aantal_personen)
    return allemaal_verschillend
```

```python
## 23 personen, 5000 herhalingen, hoeveel keer allemaal verschillend
print(f"Gemiddeld aantal pogingen met allemaal verschillende "
      f"verjaardag {np.mean(simulatie_verjaardagsparadox(23, 5000))}")
print(f"De exacte waarde is {tabel[23]}")
```
